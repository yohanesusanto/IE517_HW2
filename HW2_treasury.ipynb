```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import warnings
import pandas as pd
import numpy as np

#import data
df = pd.read_csv('Treasury-Squeeze-test.csv')
X = df.iloc[:, 2:11]
y = df.iloc[: ,-1]

print(X.shape)
print(y.shape)
```

```python
#standardizing the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
#K-Nearest-Neighbors
score = []
score1 = []
for i in range(1,100):
    knn = KNeighborsClassifier(n_neighbors=i, p=2, metric='minkowski')
    knn.fit(X_train_std, y_train)
    y_pred = knn.predict(X_test)
    score.append(knn.score(X_test,y_test)) #1st method of checking accuracy
    score1.append(accuracy_score(y_test,y_pred)) #2nd method of checking accuracy
```

```python
#confirming that both methods are equal to each other
score == score1
```

```python
#plotting the scores
plt.title('K-Nearest-Neighbors')
plt.plot(range(1,100),score,'-o')
plt.xlabel('neighbors')
plt.ylabel('accuracy')
plt.show()
```

```python
accuracy = []
for i in range(1,26):
    tree = DecisionTreeClassifier(criterion='entropy', max_depth=i, random_state=0)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    accuracy.append(accuracy_score(y_test,y_pred))
```

```python
#plotting the accuracies
plt.title("Decision Tree")
plt.plot(range(1,26),accuracy,'-o')
plt.xlabel('max-depth')
plt.ylabel('accuracy')
plt.show()
```
